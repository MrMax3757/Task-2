```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Salary Data.csv')
```

```python
df
```

```python
x=df[['YearsExperience']]
y=df[['Salary']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42) # Correct the order of variables assigned
reg = LinearRegression()
reg.fit(x_train, y_train)
predict = reg.predict(x_test)
```

```python
cd=float(input("enter your years of eperience: "))
input_data=pd.DataFrame([[cd]],columns=['YearsExperience'])
predicted_salary=reg.predict(input_data)

print(f"the predicted salary for {cd} years of exp is {predicted_salary[0][0]}")
```

```python
graph=plt.scatter(x_test,y_test,color='black')
plt.plot(x_test,predict,color='green')
```

```python
#kmeans clustering
import numpy as np
from sklearn.cluster import KMeans
data = np.array([
    [1, 2],
    [2, 3],
    [3, 2],
    [8, 7],
    [9, 8],
    [8, 9],
    [1, 4],
    [5, 9],
    [1, 9],
    [8, 7]
])

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')
plt.show()
```
